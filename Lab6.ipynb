## 12.2 Monthly Sales
Create a program that allows you to view and edit the sales amounts for each month of the current year.

### Console:
```powershell

COMMAND MENU
view	- View sales for specified month 
edit	- Edit sales for specified month 
totals - View sales summary for year

Command: edit

Command: totals
Yearly total: 67,855.00

Command: view
Three-letter Month: july 
Invalid three-letter month.

Command: exit Bye!
```

### Specifications:
- Use the attached text file named `monthly_sales.txt` that consists of rows that contain three-letter abbreviations for the month and the monthly sales.
- The program should read the file and store the sales data for each month in a dictionary with the month abbreviation as the key for each item.
- Whenever the sales data is edited, the program should write the changed data to the text file.




````python
print("Monthly Sales program\n")

MONTHS = {
    "jan", "feb", "mar", "apr", "may", "jun",
    "jul", "aug", "sep", "oct", "nov", "dec"
}

def display_menu():
    print("COMMAND MENU")
    print("view\t- View sales for specified month")
    print("edit\t- Edit sales for specified month")
    print("totals\t- View sales summary for year")
    print("exit\t- Exit program")

def get_month():
    month = input("Three-letter Month: ").strip().lower()
    if month not in MONTHS:
        print("Invalid three-letter month.")
        return ""
    return month

def view(sales):
    month = get_month()
    if month != "":
        print(f"Sales amount for {month.title()} is {sales[month]:,.2f}.")

def edit(sales):
    month = get_month()
    if month == "":
        return

    amount_input = input("Sales Amount: ").strip()

    amount_check = amount_input.replace(".", "", 1)
    if not amount_check.isdigit():
        print("Invalid sales amount.")
        return

    amount = float(amount_input)
    sales[month] = amount

    with open("monthly_sales.txt", "w") as file:
        for m, amt in sales.items():
            file.write(f"{m.title()}\t{int(amt)}\n")

    print(f"Sales amount for {month.title()} is {amount:,.2f}.")

def totals(sales):
    total = sum(sales.values())
    average = total / len(sales)

    print(f"Yearly total: {total:,.2f}")
    print(f"Monthly average: {average:,.2f}")

def main():
    display_menu()

    sales = {}
    with open("monthly_sales.txt", "r") as file:
        for line in file:
            month, amount = line.strip().split("\t")
            sales[month.lower()] = float(amount)

    while True:
        command = input("\nCommand: ").strip().lower()

        if command == "view":
            view(sales)
        elif command == "edit":
            edit(sales)
        elif command == "totals":
            totals(sales)
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Invalid command.")

if __name__ == "__main__":
    main()
````
