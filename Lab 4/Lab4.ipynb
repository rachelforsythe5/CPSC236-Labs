## 6.1 Prime Number Checker
Create a program that checks whether a number is a prime number and displays its factors if it is not a prime number.

### Console:
```powershell
Prime Number Checker

Please enter an integer between 1 and 5000: 5
5 is a prime number.

Try again? (y/n): y

Please enter an integer between 1 and 5000: 6
6 is NOT a prime number.
It has 4 factors: 1 2 3 6

Try again? (y/n): y

Please enter an integer between 1 and 5000: 200
200 is NOT a prime number.
It has 12 factors: 1 2 4 5 8 10 20 25 40 50 100 200

Try again? (y/n): n

Bye!
```

### Specifications:
- A prime number is divisible by two factors (1 and itself). For example, 7 is a prime number because it is only divisible by 1 and 7.
- If the user enters an integer that's not between 1 and 5000, the program should display an error message.
- If the number is a prime number, the program should display a message.
If the number is not a prime number, the program should display a message. Then, it should display the number of factors for the number and a list of those factors.
- Store the factors for each number in a list.
- Use functions to organize the code for this program.


````python
print("Prime Number Checker")

def get_integer():
    integer = input("\nPlease enter an integer between 1 and 5000: ")

    if not integer.isdigit():
        while not integer.isdigit():
            print("Invalid integer. Please try again.")
            integer = input("\nPlease enter an integer between 1 and 5000: ")
    elif int(integer) <= 1:
        while int(integer) <= 1:
            print("Invalid integer. Please try again.")
            integer = input("\nPlease enter an integer between 1 and 5000: ")
    elif int(integer) >= 5000:
        while int(integer) >= 5000:
            print("Invalid integer. Please try again.")
            integer = input("\nPlease enter an integer between 1 and 5000: ")
    return integer

def factors(integer):
    factors_list = []
    num_factors = 0
    for j in range(1, integer+1):
        if integer % j == 0:
            num_factors += 1
            factors_list.append(j)
    return num_factors, factors_list

def main():
    choice = "y"

    while choice == "y":
        integer = get_integer()

        integer = int(integer)
        
        if integer == 2 or integer == 3:
            print(integer, "is a prime number")
            
        elif integer > 1 and integer < 5000:
            num_factors, factors_list = factors(integer)
            if len(factors_list) > 2:
                print(integer, "is NOT a prime number")
                print("It has", num_factors, "factors:", *factors_list)
            else:
                print(integer, "is a prime number.")
        
        choice = input("\nTry again? (y/n): ")
        
    print("\nBye!")

if __name__ == "__main__":
    main()
````

## 6.2 - Contact Manager
Create a program that a user can use to manage the primary email address and phone number for a contact.

### Console
```powershell
Contact Manager

COMMAND MENU
list - Display all contacts
view - View a contact
add - Add a contact
del - Delete a contact
exit - Exit program

Command: list

2. Eric Idle

Command: view

Name: Eric Idle
Email: eric@ericidle.com
Phone: +44 20 7946 0958

Command: add
Name: Mike Murach

Mike Murach was added.

Command: list
1. Eric Idle
2. Mike Murach

```

### Specifications:
- Use a list of lists to store the data for the contacts. Provide starting data for two or more contacts.
- For the `view` and `del` commands, display an error message if the user enters and invalid contact number.
- When you exit the program, all changes that you made to the contact list are lost.




````python
print("Contact Manager\n")

def list(manager_list):
    if len(manager_list) == 0:
        print("There are no managers in the list.\n")
        return
    else:
        i = 1
        for row in manager_list:
            print(str(i) + ". " + row[0])
        print()

def view(manager_list):
    number = int(input("Number: "))
    if number < 1 or number > len(manager_list):
        print("Invalid manager number.\n")
    else:
        print("Name: " + manager_list[number-1][0])
        print("Email: " + manager_list[number-1][1])
        print("Phone: " + manager_list[number-1][2])

def add(manager_list):
    name = input("Name: ")
    email = input("Email: ")
    phone = input("Phone: ")
    manager = []
    manager.append(name)
    manager.append(email)
    manager.append(phone)
    manager_list.append(manager)
    print(manager[0] + " was added.\n")

def delete(manager_list):
    number = int(input("Number: "))
    if number < 1 or number > len(manager_list):
        print("Invalid manager number.\n")
    else:
        manager = manager_list.pop(number-1)
        print(manager[0] + " was deleted.\n")

def display_menu():
    print("COMMAND MENU")
    print("list - Display all contacts")
    print("view - View a contact")
    print("add - Add a contact")
    print("delete - Delete a contact")
    print("exit - Exit program")
    print()

def main():
    manager_list = [["Guido van Rossum", "123-456-7890", "guido@gmail.com"],
                    ["Eric Idle", "+44 20 7946 0958", "eric@ericidle.com"]
    ]
    
    display_menu()
    
    while True:
        command = input("\nCommand: ")
        if command == "list":
            list(manager_list)
        elif command == "view":
            view(manager_list)
        elif command == "add":
            add(manager_list)
        elif command == "delete":
            delete(manager_list)
        elif command == "exit":
            break
        else:
            print("Not a valid command. Please try again.\n")
            
    print("Bye!")

if __name__ == "__main__":
    main()
````
