## 3.1 Letter Grade Converter
Create a program that converts number grades to letter grades.

### Console:
```powershell
Letter Grade Converter

Enter numerical grade: 90 
Letter grade: A

Continue? (y/n): y

Enter numerical grade: 88 
Letter grade: A

Continue? (y/n): y

Enter numerical grade: 80 
Letter grade: B

Continue? (y/n): y

Enter numerical grade: 67 
Letter grade: C

Continue? (y/n): y

Enter numerical grade: 59 
Letter grade: F

Continue? (y/n): n

Bye!
```

### Specifications:
- Assume that the user will enter valid integers for the grades.
- The program should continue only if the user enters “y” or “Y” to continue.
- The grading criteria is as follows:

| Letter |	Range |
|---|-------|
| A |	88-100 |
| B |	80-87 |
| C |	67-79 |
| D |	60-66 |
| F |	<60 |

````python
print('Letter Grade Converter')

choice = 'y'
while choice.lower() == 'y':
    numGrade = int(input('\nEnter numerical grade: '))
    
    if numGrade >= 88:
        grade = 'A'
    elif numGrade >= 80:
        grade = 'B'
    elif numGrade >= 67:
        grade = 'C'
    elif numGrade >= 60:
        grade = 'D'
    else:
        grade = 'F'
    print('Letter grade: ' + grade)

    choice = input('\nContinue? (y/n): ')
print('\nBye!')
````

## 3.2 - Tip Calculator
Create a program that calculates three options for an appropriate tip to leave after a meal at a restaurant.

### Console
```powershell

Tip amount:	7.85
Total amount: 60.16

20%
Tip amount:	10.46

```

### Specifications:
- The program should calculate and display the cost of tipping at 15%, 20%, or 25%.
- Assume the user will enter valid data.
- The program should round results to a maximum of two decimal places.



````python
print('Tip Calculator')

cost = float(input('\nCost of meal: '))

# Calculating 15% tip on meal

tip15 = cost * .15
total15 = cost + tip15

print('\n15%')
print('Tip amount:\t', round(tip15, 2))
print('Total amount:\t', round(total15, 2))

# Calculating 20% tip on meal

tip20 = cost * .20
total20 = cost + tip20

print('\n20%')
print('Tip amount:\t', round(tip20, 2))
print('Total amount:\t', round(total20, 2))

# Calculating 25% tip on meal

tip25 = cost * .25
total25 = cost + tip25

print('\n25%')
print('Tip amount:\t', round(tip25, 2))
print('Total amount:\t', round(total25, 2))
````

## 3.3 - Change Calculator
Create a program that calculates the coins needed to make change for the specified number of cents.

### Console
```powershell
Change Calculator

Enter number of cents (0-99): 99

Quarters: 3
Dimes:	2
Nickels: 0
Pennies: 4

Continue? (y/n): y

Enter number of cents (0-99): 55

Quarters: 2
Dimes:	0
Nickels: 1
Pennies: 0

Continue? (y/n): n 

Bye!
```

### Specifications
- The program should display the minimum number of quarters, dimes, nickels, and pennies that one needs to make up the specified number of cents.
- Assume that the user will enter a valid integer for the number of cents.
- The program should continue only if the user enters “y” or “Y” to continue.


````python
print('Change Calculator')

choice = 'y'

while choice.lower() == 'y':
    cents = int(input('\nEnter number of cents (0-99): '))

    while cents != 0:
        # Calculating num of quarters
        quarters = cents // 25
        cents -= quarters * 25

        # Calculating num of dimes
        dimes = cents // 10
        cents -= dimes * 10

        # Calculating num of nickels
        nickels = cents // 5
        cents -= nickels * 5

        # Calculating num of pennies
        pennies = cents // 1
        cents -= pennies

    print('\nQuarters:', quarters)
    print('Dimes:', dimes)
    print('Nickels:', nickels)
    print('Pennies:', pennies)

    choice = input('\nContinue? (y/n): ')
print('\nBye!')
````

## 3.4 - Table of Powers
Create a program that displays a table of squares and cubes for the specified range of numbers.

### Console
```powershell
Table of Powers

Start number: 90
Stop number: 100

Number	Squared	Cubed
======	=======	=====
90      8100	729000
91      8281	753571
92      8464	778688
93      8649	804357
94      8836	830584
95      9025	857375
96      9216	884736
97      9409	912673
98      9604	941192
99      9801	970299
100     10000	1000000
```

### Specifications
- The formulas for calculating squares and cubes are:

    `square = x ** 2`
    
    `cube = x ** 3`

- Use tabs to align the columns.
- Assume that the user will enter valid integers.
- Make sure the user enters a start integer that’s less than the stop integer. If the user enters a start integer that’s greater than the stop integer, display an error message and give the user a chance to enter the integers again.


````python
print('Table of Powers')

start = int(input('\nStart number: '))
stop = int(input('Stop number: '))

print('\nNumber\tSquared\tCubed')
print('======\t=======\t=====')
count = start
while count != stop+1:
    square = count ** 2
    cube = count ** 3
    print(count, '\t', square, '\t', cube, '\t', sep='')
    count += 1
````
